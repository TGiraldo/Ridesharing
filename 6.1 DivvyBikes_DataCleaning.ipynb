```python
#Create trip duration in mins/ hours to add to df
df['trip_duration'] = df['ended_at'] - df['started_at']
```

```python
#Create distance traveled based on geographical coordinates 

from geopy.distance import geodesic

# Define a function to calculate Vincenty distance
def vincenty_distance(row):
    if '0.0' in (row['start_lat'], row['start_lng'], row['end_lat'], row['end_lng']):
        return None  
    
    start_coords = (row['start_lat'], row['start_lng'])
    end_coords = (row['end_lat'], row['end_lng'])
    return geodesic(start_coords, end_coords).miles

# Apply the function to create a new 'trip_distance' column
df['trip_distance_miles'] = df.apply(vincenty_distance, axis=1)

# Display the DataFrame with the new 'trip_distance' column
print(df[['start_lat', 'start_lng', 'end_lat', 'end_lng', 'trip_distance_miles']])
```
